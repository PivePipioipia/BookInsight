### Item Metadata
- **main_category** *(str)*: Main category (i.e., domain) of the product.  
- **title** *(str)*: Name of the product.  
- **average_rating** *(float)*: Rating of the product shown on the product page.  
- **rating_number** *(int)*: Number of ratings in the product.  
- **features** *(list)*: Bullet-point format features of the product.  
- **description** *(list)*: Description of the product.  
- **price** *(float)*: Price in US dollars (at time of crawling).  
- **images** *(list)*: Images of the product (sizes: thumb, large, hi_res).  
  - The `variant` field shows the position of image.  
- **videos** *(list)*: Videos of the product including title and url.  
- **store** *(str)*: Store name of the product.  
- **categories** *(list)*: Hierarchical categories of the product.  
- **details** *(dict)*: Product details, including materials, brand, sizes, etc.  
- **parent_asin** *(str)*: Parent ID of the product.  
- **bought_together** *(list)*: Recommended bundles from the websites.

```python
import re

def process_subtitle(x):
    if not isinstance(x, str) or x.strip() == "":
        return "Unknown", ""

    # Format detection
    if "Kindle" in x:
        fmt = "Kindle"
    elif "Paperback" in x:
        fmt = "Paperback"
    elif "Hardcover" in x:
        fmt = "Hardcover"
    elif "Unknown Binding" in x:
        fmt = "Unknown"
    else:
        fmt = "Unknown"

    # Edition info
    # (1) Nếu có số edition
    match = re.search(r"(\d+(st|nd|rd|th) Edition)", x, re.IGNORECASE)
    if match:
        edition = match.group(1)
    # (2) Nếu có dấu gạch ngang "–", lấy phần sau
    elif "–" in x:
        edition = x.split("–", 1)[1].strip()
    else:
        edition = ""

    return fmt, edition

df[["format", "edition_info"]] = df["subtitle"].apply(
    lambda x: pd.Series(process_subtitle(x))
)

print(df[["subtitle", "format", "edition_info"]].head(15))
```

**Các bước xử lý đề ra:**


1. main_category (str): Danh mục chính (domain) của sản phẩm.
    Đa phần là "Books" 900/1000, ngoài ra còn có "Buy a Kindle", "Audible Audiobooks"
   -> giữ lại để lọc 

2. title (str): Tên sách
    Duy nhất, quan trọng
   -> giữ 

3. average_rating (float): Điểm trung bình của sản phẩm hiển thị trên trang Amazon.

   Numeric float (1-5). Dùng để filter, rank
   -> giữ 

4. rating_number (int): Số lượng đánh giá (rating count).

   Số lượng đánh giá, để tính đột in cậy của rating

   -> giữ 

5. features (list): Danh sách đặc điểm nổi bật của sản phẩm (bullet points).

   Nối list thành text, dùng trong combined_text

   -> features = [] → cần convert thành chuỗi rỗng ""

   -> chắc chắn giữ 

7. description (list): Mô tả chi tiết sản phẩm.

   text dài, giàu thông tin, rất hữu ích

   -> description = [] → cần convert thành chuỗi rỗng ""

   -> nối toàn bộ list lại thành text

8. price (float): Giá bằng USD (tại thời điểm crawl).

   -> convert sang numeric (float)

   -> xử lý giá trị lỗi, thiếu thông tin

   -> xử lý outliers

9. images (list): Ảnh sản phẩm (có nhiều kích thước: thumbnail, large, hi_res). Trường variant cho biết vị trí ảnh.

   -> chỉ giữ large để hiển thị ảnh, còn lại trống

   -> tạo main_images chỉ chứa link ảnh (large)
    

10. videos (list): Video sản phẩm, bao gồm tiêu đề và đường dẫn.

    -> link trống -> drop

    

11. store (str): Tên cửa hàng bán sản phẩm.

    -> chứa các thông tin như author, edit,... không phải là cửa hàng, dư thừa và không đủ

    -> xóa cột store

12. categories (list): Danh mục phân cấp của sản phẩm.

    -> xóa cột đầu : "Books"

13. details (dict): Chi tiết sản phẩm (vật liệu, thương hiệu, kích cỡ, …).

    -> giữ 

14. parent_asin (str): Mã định danh cha của sản phẩm.

    -> giữ lấy tham chiếu 

15. bought_together (list): Các gợi ý mua kèm từ website.

    -> drop cột này

16. subtitle : mô tả định dạng sách

    -> giữ và tách ra các cột phụ

    -> Tạo cột format (Kindle, Paperback, Hardcover, Unknown Binding…).

    -> Tạo cột edition_info (số edition hoặc ngày xuất bản nếu có).

18. author

    -> keep, tách thành các cột phụ
